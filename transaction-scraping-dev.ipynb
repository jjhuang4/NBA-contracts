```python
from selenium import webdriver
from selenium.common import ElementClickInterceptedException, StaleElementReferenceException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import numpy as np
import os
import re
import time
from datetime import date, datetime
```

```python
def load_transactions_till_date(driver, stop_date):
    try:
        # replace 'closeButton' with the actual ID or class name of the close button
        closePopupButton = WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.CLASS_NAME, 'mc-closeModal')))
        closePopupButton.click()
    except:
        print("No popup found.")
    
    while True:
        try: 
            #dates = [date.text for date in WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'tbody tr td h3')))]
            WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'tbody tr td span a')))
            WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'tbody tr td h3')))
            dates = [date.text for date in driver.find_elements(By.CSS_SELECTOR, 'tbody tr td h3')]
            flag = 0
            for date in dates:
                try:
                    date = datetime.strptime(date, '%b %d %Y').date()
                    print(date)
                except Exception as e:
                    continue
                # print(date)
                if date < stop_date:
                    print("Finished loading until stop-date successfully")
                    flag = 1
                    break
            if flag == 1:
                break
                
            # Click on view more transactions button to show more elements
            button = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.CLASS_NAME, 'show-more')))
            driver.execute_script("arguments[0].scrollIntoView();", button)
            try:
                button.click()
            except ElementClickInterceptedException as e:
                print(e)
                print("Executing JavaScript to click:")
                driver.execute_script("arguments[0].click();", button)
        except StaleElementReferenceException:
            continue  # DOM changed, retry fetching elements
```

```python
wait = 1
# "Jun 30 2022"
stop_date = datetime.strptime("Feb 2 2024", '%b %d %Y').date()
print(stop_date)

driver = webdriver.Chrome()
time.sleep(wait)
driver.get(f'https://www.spotrac.com/nba/transactions/trade/')
driver.implicitly_wait(10)
print("successful entrance\n\n")

# load_transactions_till_date(driver, stop_date)
# WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'tbody tr')))
try:
    # replace 'closeButton' with the actual ID or class name of the close button
    closePopupButton = WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.CLASS_NAME, 'mc-closeModal')))
    closePopupButton.click()
except:
    print("No popup found.")

# Main scraping code block
# --------------------------------------------------------------------------------------------------------------
data = []
flag = 0
while True:
    try:
        # Explicit wait for elements
        WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'tbody tr td h3')))
        WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'tbody tr td span span a')))
        # Fetching and parsing through elements
        rows = driver.find_elements(By.TAG_NAME, 'table')
        for row in rows:
            #print(row.text)
            date, acquires = [part for part in row.find_elements(By.CSS_SELECTOR, 'tbody')]
            date = date.text
            try:
                date = datetime.strptime(date, '%b %d %Y').date()
            except Exception as e:
                print("Date not loaded, continuing")
                continue
            # check if date is less than or equal to stop date, in which case we stop execution
            if date < stop_date:
                flag = 1
                break
            print("Date: ")
            print(date)
            print("\n")
            # team = "PLACEHOLDER"
            trade_data = acquires.find_elements(By.CLASS_NAME, 'tradedata')
            for team_data in trade_data:
                team = team_data.find_element(By.CLASS_NAME, 'tradelabel')
                team = team.text.split(' ')
                team = ' '.join(team[0:len(team)-1])
                # players = team_data.find_elements(By.TAG_NAME, 'span a')
                # for player in players:
                #     print(f"{player.text} acquired by {team}")
                tradeplayers = team_data.find_elements(By.CLASS_NAME, 'tradeplayer')
    
                for trade in tradeplayers:  # each trade a single span line with class tradeplayer
                    try:
                        player = trade.find_element(By.TAG_NAME, 'a').text.rsplit(' ', 1)
                        name = player[0]
                        position = player[1].replace("(", "").replace(")", "")
                        salary = ""
                        if player is not None:
                            content = trade.get_attribute('innerHTML')
                            parts = content.split('(')
                            salary = parts[-1].strip(')')
                        if salary.startswith('$'):
                            salary = int(salary.replace('$', '').replace(',',''))
                        else:
                            salary = np.nan
                        print(f"{name}, {position}, {salary}, acquired by {team}")
                        data.append([date, name, position, salary, team])
                    except Exception as e:  # error occurred from not being able to find 'a' tag
                        continue
            print("--------------------------------------------------------------")
        
        if flag == 1:
            print("Finished scraping until stop-date successfully")
            break
         # Click on view more transactions button to show more elements
        button = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.CLASS_NAME, 'show-more')))
        driver.execute_script("arguments[0].scrollIntoView();", button)
        try:
            button.click()
        except ElementClickInterceptedException as e:
            print(e)
            print("Executing JavaScript to click:")
            driver.execute_script("arguments[0].click();", button)
    except StaleElementReferenceException as e:
        continue
    except Exception as e:
        print("Encountered unexpected error, aborting")
        print(e)
        break
driver.quit()
# --------------------------------------------------------------------------------------------------------------
```

```python
wait = 1
# "Jun 30 2022"
stop_date = datetime.strptime("Jun 30 2022", '%b %d %Y').date()
print(stop_date)

driver = webdriver.Chrome()
time.sleep(wait)
driver.get(f'https://www.spotrac.com/nba/transactions/trade/')
driver.implicitly_wait(10)
print("successful entrance\n\n")

# load_transactions_till_date(driver, stop_date)
# WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'tbody tr')))
try:
    # replace 'closeButton' with the actual ID or class name of the close button
    closePopupButton = WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.CLASS_NAME, 'mc-closeModal')))
    closePopupButton.click()
except:
    print("No popup found.")

# Main scraping code block
# --------------------------------------------------------------------------------------------------------------
data = []
flag = 0
while True:
    try:
        # Explicit wait for elements
        WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'tbody tr td h3')))
        WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'tbody tr td span span a')))
        # Fetching and parsing through elements
        tables = driver.find_elements(By.CSS_SELECTOR, 'tbody')
        for table in tables:
            #print(row.text)
            #print(table.text)
            date, acquires = [part for part in table.find_elements(By.CSS_SELECTOR, 'tbody tr')]
            date = date.text
            try:
                date = datetime.strptime(date, '%b %d %Y').date()
            except Exception as e:
                print("Date not loaded, continuing")
                continue
            # check if date is less than or equal to stop date, in which case we stop execution
            if date < stop_date:
                flag = 1
                break
            print("Date: ")
            print(date)
            print("\n")
            team = "PLACEHOLDER"
            team_acquisitions = acquires.find_elements(By.CSS_SELECTOR, 'tbody, tr, td')
            for team_transaction in team_acquisitions:
                trade_data = team_transaction.find_elements(By.CSS_SELECTOR, 'tbody tr td span span')
                # for team_data in trade_data:
                team = trade_data[0].text
                team = team.split(' ')
                team = ' '.join(team[0:len(team) - 1])
                # players = team_data.find_elements(By.TAG_NAME, 'span a')
                # for player in players:
                #     print(f"{player.text} acquired by {team}")
                tradeplayers = trade_data[1:]
    
                for trade in tradeplayers:  # each trade a single span line with class tradeplayer
                    try:
                        if "round pick" in trade.text: # if statement filters out non-player entries
                            non_player_entry = trade.text
                            data.append([date, non_player_entry, None, np.nan, team, "PICK"])
                            print(f"{non_player_entry} acquired by {team}")
                            continue
                        name, position = "", ""
                        if len(trade.text.split(' ')) > 3:
                            player = trade.text.rsplit(' ', 1)
                            # print(player)
                            player_pos = player[0].rsplit(' ', 1)
                            name = ' '.join(player_pos[:-1])
                            position = player_pos[-1].replace("(", "").replace(")", "")
                        else:
                            player = trade.text.rsplit(' ', 1)
                            name = player[0]
                            position = player[1].replace("(", "").replace(")", "")
                        salary = ""
                        if player is not None:
                            content = trade.get_attribute('innerHTML')
                            parts = content.split('(')
                            salary = parts[-1].strip(')')
                        if salary.startswith('$'):
                            salary = int(salary.replace('$', '').replace(',', ''))
                        else:
                            salary = np.nan
                        print(f"{name}, {position}, {salary}, acquired by {team}")
                        data.append([date, name, position, salary, team, "PLAYER"])
                    except Exception as e:  # error occurred from not being able to find 'a' tag
                        continue
            print("--------------------------------------------------------------")
        if flag == 1:
            print("Finished scraping until stop-date successfully")
            break
        # Click on view more transactions button to show more elements
        button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CLASS_NAME, 'show-more')))
        driver.execute_script("arguments[0].scrollIntoView();", button)
        try:
            button.click()
        except ElementClickInterceptedException as e:
            print(e)
            print("Executing JavaScript to click:")
            driver.execute_script("arguments[0].click();", button)
    except StaleElementReferenceException as e:
        continue
    # except Exception as e:
    #     print("Encountered unexpected error, aborting")
    #     print(e)
    #     break
driver.quit()
```

```python
df = pd.DataFrame(data, columns=['Date', 'Name', 'Position', 'Payment', 'Team', 'Type'])
```
